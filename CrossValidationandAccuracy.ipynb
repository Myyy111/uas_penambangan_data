```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
```

```python
# 1. Memuat Dataset
df = pd.read_csv('sample_data/databeasiswa.csv')
```

```python
# 2. Menentukan Fitur (X) dan Target (y)
# Kita menggunakan IPK dan Tingkat Kemiskinan sebagai fitur prediktor
X = df[['IPK', 'Tingkat Kemiskinan (TM)']]
y = df['Beasiswa/Tidak *']
```

```python
# 3. Inisialisasi Model
# Menggunakan Random Forest sebagai algoritma klasifikasi
model = RandomForestClassifier(random_state=42)
```

```python
#4. Konfigurasi Cross-Validation (K-Fold)
# n_splits=5 berarti data dibagi menjadi 5 bagian (fold)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# 5. Menjalankan Cross-Validation
scores = cross_val_score(model, X, y, cv=kf)
```

```python
# 6. Menampilkan Hasil
print("--- Hasil Cross-Validation ---")
for i, score in enumerate(scores, 1):
    print(f"Akurasi Fold-{i}: {score:.4f}")

print("-" * 30)
print(f"Rata-rata Akurasi: {scores.mean():.4f}")
print(f"Standar Deviasi: {scores.std():.4f}")
```
